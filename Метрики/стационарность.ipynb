```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.stattools import adfuller
import statsmodels.api as sm
from sklearn.linear_model import LassoCV
from sklearn.preprocessing import StandardScaler
import warnings

warnings.filterwarnings('ignore')

plt.style.use('seaborn-v0_8-darkgrid')
plt.rcParams['figure.figsize'] = (14, 6)
```

```python
from read_functions import dict_func

# Собираем мастер-фрейм
all_series = []
for path, func in dict_func.items():
    try:
        df = func(path)
        df['date'] = pd.to_datetime(df['date'])
        all_series.append(df.set_index('date'))
    except Exception as e:
        print(f"Ошибка загрузки {path}: {e}")

df_raw = pd.concat(all_series, axis=1)
df_raw = df_raw.sort_index()

print(f"Сырые данные загружены. Диапазон: с {df_raw.index.min().date()} по {df_raw.index.max().date()}")
print(f"Размерность: {df_raw.shape}")
df_raw.head()
```

```python
# Ищем точку отсчета по сырым данным 
first_loan_date = df_raw['loans_kz_total'].dropna().index.min()
print(f"Старт выборки: {first_loan_date.date()}")

# Создаем  ежедневный календарь
daily_index = pd.date_range(start='2020-01-01', end='2026-01-01', freq='D')
df_processed = pd.DataFrame(index=daily_index).join(df_raw)

# Логарифмируем кредиты
loan_cols = [c for c in df_processed.columns if 'loans_kz' in c]
for col in loan_cols:
    df_processed[col] = np.log1p(df_processed[col])

# Разделяем признаки по логике заполнения
step_cols = ['macro_kz_base_rate', 'macro_ru_key_rate']
daily_market_cols = [
    c for c in df_processed.columns 
    if c.startswith('bond_') or c.startswith('brent') or c == 'KASE_BMC' or c.startswith('macro_ru_ruonia')
]
macro_interp_cols = [
    c for c in df_processed.columns 
    if c not in step_cols + daily_market_cols
]

df_processed[step_cols + daily_market_cols] = df_processed[step_cols + daily_market_cols].ffill()
df_processed[macro_interp_cols] = df_processed[macro_interp_cols].interpolate(method='linear')

df_processed = df_processed.ffill().bfill()

print(f"Размерность идеальной матрицы уровней: {df_processed.shape}")
df_processed.isna().sum().sum()
```

```python
def adf_test(series, signif=0.05):
    series_clean = series.dropna()
    if len(series_clean) < 30: return None, False
    try:
        res = adfuller(series_clean, autolag='AIC')
        return res[1], res[1] < signif
    except Exception:
        return None, False

df_stationary = pd.DataFrame(index=df_processed.index)
report = []

for col in df_processed.columns:
    raw_series = df_processed[col]
    
    # Тест Уровней
    p_val_0, is_stat_0 = adf_test(raw_series)
    
    if is_stat_0:
        df_stationary[col] = raw_series
        report.append({'Метрика': col, 'Порядок': 'I(0)', 'Действие': 'Без изменений', 'p-value': round(p_val_0, 4)})
    else:
        # Тест Первой разности
        diff1 = raw_series.diff()
        p_val_1, is_stat_1 = adf_test(diff1)
        
        if is_stat_1:
            df_stationary[col] = diff1
            report.append({'Метрика': col, 'Порядок': 'I(1)', 'Действие': '1-я разность', 'p-value': round(p_val_1, 4)})
        else:
            # Вторая разность
            diff2 = diff1.diff()
            df_stationary[col] = diff2
            report.append({'Метрика': col, 'Порядок': 'I(2)+', 'Действие': '2-я разность', 'p-value': '...'})

report_df = pd.DataFrame(report)
print("\n=== Отчет о стационарности ===")
with pd.option_context('display.max_rows', None):
    display(report_df)

# Из-за .diff() первая строчка (или две) станут NaN. Дропаем только их.
df_stationary_clean = df_stationary.dropna(how='all').dropna()
df_stationary_clean.to_csv('master_df_daily_stationary.csv')
print(f"\nСтационарная матрица готова и сохранена. Размерность: {df_stationary_clean.shape}")
```

```python
df_raw_filtered = df_raw.loc['2020-01-01':'2026-01-01']

df_monthly_raw = df_raw_filtered.resample('ME').mean()

step_cols = ['macro_kz_base_rate', 'macro_ru_key_rate']
daily_market_cols = [
    c for c in df_monthly_raw.columns 
    if c.startswith('bond_') or c.startswith('brent') or c == 'KASE_BMC' or c.startswith('macro_ru_ruonia')
]
macro_interp_cols = [
    c for c in df_monthly_raw.columns 
    if c not in step_cols + daily_market_cols
]

df_monthly_raw[step_cols + daily_market_cols] = df_monthly_raw[step_cols + daily_market_cols].ffill()
df_monthly_raw[macro_interp_cols] = df_monthly_raw[macro_interp_cols].interpolate(method='linear')

df_monthly_processed = df_monthly_raw.ffill().bfill()

df_monthly_stationary = pd.DataFrame(index=df_monthly_processed.index)
monthly_report = []

for col in df_monthly_processed.columns:
    if 'loans_kz' in col:
        raw_series = np.log1p(df_monthly_processed[col])
    else:
        raw_series = df_monthly_processed[col]
        
    # Тест Уровней 
    p_val_0, is_stat_0 = adf_test(raw_series)
    
    if is_stat_0:
        df_monthly_stationary[col] = raw_series
        monthly_report.append({'Метрика': col, 'Порядок': 'I(0)', 'Действие': 'Без изменений', 'p-value': round(p_val_0, 4)})
    else:
        # Тест Первой разности
        diff1 = raw_series.diff()
        p_val_1, is_stat_1 = adf_test(diff1)
        
        if is_stat_1:
            df_monthly_stationary[col] = diff1
            monthly_report.append({'Метрика': col, 'Порядок': 'I(1)', 'Действие': '1-я разность', 'p-value': round(p_val_1, 4)})
        else:
            # Вторая разность
            diff2 = diff1.diff()
            df_monthly_stationary[col] = diff2
            monthly_report.append({'Метрика': col, 'Порядок': 'I(2)+', 'Действие': '2-я разность', 'p-value': '...'})

monthly_report_df = pd.DataFrame(monthly_report)

print("=== Отчет о стационарности (МЕСЯЧНАЯ гранулярность, корректный пайплайн) ===")
with pd.option_context('display.max_rows', None):
    display(monthly_report_df)

i2_metrics_monthly = monthly_report_df[monthly_report_df['Порядок'] == 'I(2)+']['Метрика'].tolist()
df_clean = df_monthly_stationary.drop(columns=[col for col in i2_metrics_monthly if col in df_monthly_stationary.columns]).dropna()

targets = [col for col in df_clean.columns if col.startswith('bond_ru_ofz_yield_')]
features = [col for col in df_clean.columns if col not in targets]

print(f"\nСтрок для обучения (МЕСЯЦЫ): {len(df_clean)}")
print(f"Таргетов (Y): {len(targets)}")
print(f"Признаков (X): {len(features)}")
print(f"Отброшено I(2)+ признаков: {len(i2_metrics_monthly)}\n")
```

```python
results_ols = []

for target in targets:
    Y = df_clean[target]
    X = df_clean[features]
    X = sm.add_constant(X)
    
    try:
        model = sm.OLS(Y, X).fit()
        pvals = model.pvalues.drop('const')
        
        sig_features = pvals[pvals < 0.05].sort_values()
        top_factors_str = ", ".join([f"{feat} (p={pval:.3f})" for feat, pval in list(sig_features.items())[:4]])
        
        results_ols.append({
            'Срок до погашения': target,
            # 'R²': model.rsquared,
            'Adj. R²': model.rsquared_adj,
            'Топ значимых предикторов (OLS)': top_factors_str if top_factors_str else "Нет значимых факторов"
        })
    except Exception as e:
        results_ols.append({
            'Облигация (Таргет)': target,
            # 'R²': 0,
            'Adj. R²': 0,
            'Топ значимых предикторов (OLS)': f"Ошибка: {e}"
        })

res_ols_df = pd.DataFrame(results_ols).set_index('Срок до погашения')
with pd.option_context('display.max_rows', None, 'display.max_colwidth', None):
    display(res_ols_df.style.background_gradient(subset=['Adj. R²'], cmap='Blues', vmin=0, vmax=1).set_table_attributes('style="width: 70% !important; margin-right: auto;"'))
```
